The heatmap indicates that there is multicollinearity problem in the data. Multicollinearity refers to the fact that predictor varibles are strongly correlated to one another.
This is a problem in linear regression. We need predictor variables are indenpendent of one another so that the predicting model can yield optimal results. 

One method we can use to fix multicollinearity issue is using RidgeModel.
