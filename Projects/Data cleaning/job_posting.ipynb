```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
job_posting=pd.read_csv("Uncleaned_DS_jobs.csv",index_col=0)
```

```python
#Skim through data
job_posting.describe()
job_posting.info()
job_posting.head()
```

### Split and transform salary into integer

```python
job_posting['Salary Estimate'].unique()
```

```python
job_posting['Salary Estimate'].apply(lambda x: x.split('(')[0]).unique()
```

```python
job_posting['Salary Estimate']=job_posting['Salary Estimate'].apply(lambda x: x.split('(')[0])
```

```python
job_posting['Salary Estimate']=job_posting['Salary Estimate'].apply(lambda x: x.replace('K','').replace('$',''))
```

```python
job_posting['Minimum salary (in thousand)']=job_posting['Salary Estimate'].apply(lambda x: x.split('-')[0])
job_posting['Maximum salary (in thousand)']=job_posting['Salary Estimate'].apply(lambda x: x.split('-')[1])
```

```python
job_posting.head()
```

```python
job_posting['Minimum salary (in thousand)']=job_posting['Minimum salary (in thousand)'].apply(int)
```

```python
job_posting['Maximum salary (in thousand)']=job_posting['Maximum salary (in thousand)'].apply(int)
```

```python
job_posting['Maximum salary (in thousand)']
```

### Remove number from the company name

```python
job_posting['Company Name']=job_posting['Company Name'].apply(lambda x: x.split('\n')[0])
```

```python
job_posting['State']=job_posting['Location'].apply(lambda x: x.split(',')[-1])
```

```python
job_posting.head()
```

### Sorting and Creating new columns

```python
# Create column Average salary
job_posting['Average_salary']=(job_posting['Minimum salary (in thousand)'] + job_posting['Maximum salary (in thousand)'])/2
```

```python
job_posting.sort_values(by="Rating", ascending=True)
```

```python
job_posting[job_posting['Rating']==-1.0].shape
```

```python
# Replace Rating with value -1 with 0
job_posting['Rating']=np.where(job_posting.Rating == -1, 0, job_posting.Rating)
```

```python
job_posting.sort_values(by='Rating',ascending=True)
```

```python
job_posting['Founded']
```

```python
job_posting[job_posting['Founded']==-1].shape

#We use this function to check how many observations have the value specified
#Results returned inform us of how many observations with value -1
```

```python
job_posting['Years_in_business']=job_posting['Founded'].apply(lambda x: x if x<1 else 2023-x)
```

```python
job_posting.head()
```

```python
job_posting.Years_in_business = np.where(job_posting['Years_in_business']==-1,0,job_posting.Years_in_business)
```

```python
job_posting.tail()
```

```python
job_posting['Job Description'][2].split('\n')
```

```python
#Count the number of people for each title

job_posting['Job Title'].value_counts()
```

```python
job_posting['State'].value_counts()
```

```python
job_posting = job_posting[~((job_posting['State']=='Remote') | (job_posting['State']=='Utah') |
                            (job_posting['State']=='New Jersey') | (job_posting['State']=='California') |
                           (job_posting['State']=='Texas'))]
```

```python
job_posting['State']=job_posting['State'].apply(lambda x: x.replace('United States', 'US'))
job_posting['State'].value_counts()
```

```python
job_posting.drop(['Founded','Competitors'], axis=1,inplace=True)
```

```python
job_posting.head()
```

```python
job_posting = job_posting.apply(lambda x: x.strip if isinstance(x,str) else x)
```
